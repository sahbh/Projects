```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the CSV file into a DataFrame
df_trip = pd.read_csv('trips.csv')
# Merge 'business' and 'month' into single column so can plot
df_trip['business-month'] = df_trip['business'] + '-' + df_trip['month'].astype(str)
# Size of graph
plt.figure(figsize=(16, 8))
# Create histogram
plt.bar(df_trip['business-month'], df_trip['count'], color='green', log=True)

# Title and axis labels, rotate axis so easy to view
plt.xlabel('Business-Month')
plt.ylabel('Trip Counts')

plt.xticks(rotation=45)  
plt.title('Trip Count per Business per Month')
plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read the CSV file into a DataFrame
df_profit = pd.read_csv('profits.csv')
# Merge 'business' and 'month' into a single column so can plot
df_profit['business-month'] = df_profit['business'] + '-' + df_profit['month'].astype(str)
# Absolute profit column
df_profit['plot_profit'] = df_profit['total_profit'].abs()
# Adjust colours for profit margins
c = ['orange' if x < 0 else 'blue' for x in df_profit['total_profit']]
# Plot size
plt.figure(figsize=(12, 8))

# Create the histogram using a logarithmic scale
plt.bar(df_profit['business-month'], df_profit['plot_profit'], color=c, log=True)

# Title and axis labels, rotate axis so easy to view
plt.xlabel('Per Business per Month')
plt.ylabel('Platform\'s Profits')

plt.xticks(rotation=45)  
plt.title('Platform\'s Profits per Business per Month (Orange = Loss, Blue = Profit)')
plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Read the CSV file into a DataFrame
df_earnings = pd.read_csv('earnings.csv')
# Merge 'business' and 'month' into one column to plot
df_earnings['business-month'] = df_earnings['business'] + '-' + df_earnings['month'].astype(str)
# Adjust size
plt.figure(figsize=(14, 8))
# Create the histogram
plt.bar(df_earnings['business-month'], df_earnings['sum(driver_total_pay)'], color='red', log=True)

# Title and axis labels, rotate axis so easy to view
plt.xlabel('Per Business per Month')
plt.ylabel('Driver\'s Earnings')

plt.xticks(rotation=45) 
plt.title('Driver\'s Earnings per Business per Month')
plt.tight_layout()  
plt.show()
```
